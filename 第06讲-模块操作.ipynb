### 例 dir 方式获取帮助


- 语法
```python
>>> dir([object])
```

- dir() 的说明
    - 不带参数时，返回当前范围内的变量、方法和定义的类型列表
    - 带参数时，返回参数的属性、方法列表
    - 如果参数包含方法__dir__()，该方法将被调用
    - 如果参数不包含__dir__()，该方法将最大限度地收集参数信息

#### 例 help 方式获取帮助


- help函数
    - 一个内置函数，用于查看函数或模块用途的详细说明



- help 与 dir 函数的区别
    - help 函数更详细
    - dir 函数只罗列属性

#### 例  ? 和 ?? 快捷方式获得帮助

- ? 用于查看详情
- ?? 查看更具体的帮助信息，如函数代码


- 语法
```python
>>> obj?
>>> ?obj
```

- 注： Jupyter Notebook 所特有

#### 例 TAB 键方式获取帮助

- 功能
    - 补全
    - 属性提示

- 语法
```python
>>> some_letters<TAB>
>>> obj.<TAB>
```

- 说明 —— 按小数点后，续按 TAB 键

### <font size=4><span id="6.4.1"> 6.4.1 calendar模块</span></font>

````python
# 快速判断是否为闰年
# 什么是闰年？
import calendar as cal

year = int(input("Enter the year "))
if cal.isleap(year) == True:
    print("%4d is a leap year."%year)
else:
    print("%4d is not a leap year."%year)
````

````python
# 打印任意年月的日历
import calendar as cal

year = 2022
month = 4
print(cal.month(year,month))
````

````python
# 改变日历中每一周的第一天
import calendar as cal

year = 2023
month = 3
c = cal.TextCalendar(cal.SUNDAY)   # calendar.TextCalendar 类
print(c.formatmonth(year,month))
````

````python
# 打印一整年日历
# 最简单的方法
import calendar as cal

year = 2023
calendar = cal.calendar(2023,w=0,l=0,c=5,m=4)
print(calendar)
````

````python
# 打印一整年日历
# 调整日历中每一周的第一天
import calendar as cal

year = 2023
c = cal.TextCalendar(cal.SUNDAY)
calendar = c.formatyear(year,w=0,l=0,c=6,m=4)
print(calendar)
````

````python
# 母亲节：每年5月的第2个星期日
import calendar as cal

for year in range(2023,2031):
    month = cal.monthcalendar(year,5)
    mom_day = month[1][6]     # month 究竟张啥样？
    print("%4d: 5月 %2d日"%(year,mom_day))
````

### <font size=4><span id="6.4.2"> 6.4.2 time模块</span></font>

 time模块中时间表现的格式主要有三种：  

- timestamp时间戳，时间戳表示的是从1970年1月1日00:00:00开始按秒计算的偏移量

- struct_time时间元组，共有九个元素。

- format time 格式化时间，已格式化的结构使时间更具可读性。包括自定义格式和固定格式。

时间格式转换图：
<img src="images/chapter04/TimeFormatTransform.png" width=500>

````python
# timestamp 时间戳格式
# 从纪元开始算起到目前多少秒
# 1970年1月1日 00:00:00 (UTC) 为 UNIX 系统纪元
import time

seconds = time.time()
print("Seconds since epoch =",seconds)    
````

````python
# struct_time 时间元组格式，
import time

seconds = time.time()
result = time.localtime(seconds)    # tm_isdst（夏令时）
# result = time.gmtime(seconds)
print("Result:",result)   
type(result)
````

````python
# 字符串格式
# %a %b %d %H:%M:%S %Y （星期几 月 日 时:分:秒 年）
import time

seconds = time.time()
localTime = time.ctime(seconds)
print("Local time:",localTime)    
````

````python
# 时间格式转换
import time

seconds = time.time()
result = time.localtime(seconds)
result = time.strftime("%H:%M:%S",result)   # 元组 --> 字符串，输入格式由strftime的第一个参数确定
print(result)
````

### <font size=4><span id="6.4.3"> 6.4.3 datetime模块</span></font>
datatime模块重新封装了time模块，提供更多接口，提供的类有：timedelta,date,datetime,time,tzinfo等

````python
# 获得今天日期
from datetime import date

today = date.today()
print(today)         # today 是什么类型？
print("今天是%d年%d月%d日" % (today.year,today.month,today.day)) 
````

````python
# 日期格式转换
from datetime import date

today = date.today()
date_str = today.strftime("%d/%m/%y")    # ("%A %d. %B %Y")
print(date_str)
````

````python
# 替换日期
# 100年前的今天
from datetime import date

today = date.today()
today = today.replace(year=today.year-200)   # today.replace(year=self.year, month=self.month, day=self.day)

CHN = ["一","二","三","四","五","六","日"]
weekday = today.weekday()
print("200年前的今天是星期" + CHN[today.weekday()])
````

````python
# 放假倒计时
from datetime import date

today = date.today()
national_day = date(today.year, 5, 1)     # 用 年，月，日 创建一个 datetime.date 对象
if national_day < today:
    national_day = national_day.replace(year=today.year+1)
time_to_national_day = abs(national_day-today)    # datetime模块中的日期可以直接加减（time模块中只能通过时间戳的加减来间接实现日期的加减）
print("距离下一个国庆节还有%d天" % time_to_national_day.days)
````

### 连续随机变量

````python
import random

# 生成0～1区间的均布随机浮点数
print(random.random())

# 生成(a, b)区间的均布随机浮点数
a, b = 10, 20
print(random.uniform(a, b))   # a 和 b 可换位置
````

````python
# 生成均布随机浮点数序列
import random
import matplotlib.pyplot as plt

a, b = 10, 20
seq = []
for i in range(1,100000):
    seq.append(random.uniform(a,b))

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(seq, bins=50, color='blue', alpha=0.7)
plt.show()
````

````python
# 生成三角分布随机浮点数序列
import random
import matplotlib.pyplot as plt

a, b, c = 10, 20, 18
seq = []
for i in range(1,100000):
    seq.append(random.triangular(a,b,c))

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(seq, bins=50, color='blue', alpha=0.7)
plt.show()
````

````python
# 生成分布随机浮点数序列
import random
import matplotlib.pyplot as plt

mu, sigma = 1, 0.1
seq = []
for i in range(1,100000):
    seq.append(random.gauss(mu,sigma))        # lognormvariate

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(seq, bins=50, color='blue', alpha=0.7)
plt.show()
````

````python
# 基本用法：cut 和 lcut
# 两种模式：精准模式（默认） 和 全模式
import jieba

seg_list = jieba.cut("经济运行总体平稳，经济结构持续优化，国内生产总值突破100万亿元。", cut_all=False)     # 精准模式
print("精准模式: " + "/".join(seg_list))

seg_list = jieba.cut("经济运行总体平稳，经济结构持续优化，国内生产总值突破100万亿元。", cut_all=True)      # 全模式
print("全模式: " + "/".join(seg_list))
````

````python
# 中文
# 比较复杂
# 如何从句子中分出词？中文不像英文，词与词之间没有空格！
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba

f = open(r'files\chapter04\shisiwu.txt', encoding='utf-8', mode='r').read()
text_after_split = jieba.cut(str(f), cut_all=False)
words = ' '.join(text_after_split)

wc = WordCloud(background_color='white',font_path='msyh.ttc') 
wc.generate(words)
plt.imshow(wc)
plt.axis("off")
plt.show()
````

````python
# 进阶：去除停用词（的，了，啊，呀……）
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba

# 读取停用词
stop_words = [line.strip() for line in open(r'files\chapter04\cn_stopwords.txt', encoding='utf-8', mode='r').readlines()]

# 对句子进行分词
text = open(r'files\chapter04\西游记.txt', encoding='utf-8', mode='r').read()
words = ''
sentence_seged = jieba.cut(text.strip(), cut_all=False)
for word in sentence_seged:
    if word not in stop_words:
        if word != '\t':
            words += word
            words += " "
            
wc = WordCloud(background_color='white',font_path='msyh.ttc') 
wc.generate(words)
plt.imshow(wc)
plt.axis("off")
plt.show()
````
